```python
import re
def toList(s):
    p0 = '\[\'|\, \'|[ ]+\'\, \'|\'\]'  # 書式
    # print('置換前: ' + s)
    s = re.sub(p0, '', s)
    # print('置換後: ' + s)
    return s.split('\'')
```

```python
l = [df.recipeIngredient,df.recipeInstructions,df.rankings]
for x in l:
    for row in x:
        row = toList(row)
```
